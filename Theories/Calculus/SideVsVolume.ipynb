# Differential equation for the volume of a cube

Let $s(t)$ be the side length of the cube (as a function of time). The volume is
$$
V(t) = s(t)^3.
$$

Differentiating with respect to $t$:
$$
\frac{dV}{dt} = 3s(t)^2\,\frac{ds}{dt}.
$$

Using $s(t) = V(t)^{1/3}$, we have $s(t)^2 = V(t)^{2/3}$, so
$$
\frac{dV}{dt} = 3V(t)^{2/3}\,\frac{ds}{dt}.
$$

Equivalently, solving for $\frac{ds}{dt}$ in terms of $V(t)$ and $\frac{dV}{dt}$:
$$
\frac{ds}{dt} = \frac{1}{3}V(t)^{-2/3}\,\frac{dV}{dt}.
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Side lengths
s = np.linspace(0.1, 10, 500)

# Small change in side length
delta_s = 0.01

# Volume change approximation: ΔV ≈ 3 s^2 Δs
delta_V = 3 * s**2 * delta_s

plt.figure()
plt.plot(s, delta_V)
plt.xlabel("Side length s")
plt.ylabel("Volume change ΔV")
plt.title("Same Δs causes much larger ΔV for bigger cubes")
plt.grid(True)
plt.show()
```

```python
# Volume from side length
V = s**3

# Assume volume is increasing at a constant rate
dV_dt = np.ones_like(s)

# Raw (misleading) interpretation
raw_effect = dV_dt

# Corrected side-length rate
ds_dt = (1/3) * V**(-2/3) * dV_dt

plt.figure()
plt.plot(s, raw_effect, label="Raw dV/dt")
plt.plot(s, ds_dt, label="Corrected ds/dt using V^(-2/3)")
plt.xlabel("Side length s")
plt.ylabel("Rate")
plt.title("V^{-2/3} removes geometric amplification")
plt.legend()
plt.grid(True)
plt.show()
```
