### First Derivative

**2-point formulas:**
$$
\begin{aligned}
\text{Forward:} \quad f'(x_0) &\approx \frac{f_1 - f_0}{h} \\
\text{Backward:} \quad f'(x_0) &\approx \frac{f_0 - f_{-1}}{h}
\end{aligned}
$$

**3-point formulas:**
$$
\begin{aligned}
\text{Central:} \quad f'(x_0) &\approx \frac{f_1 - f_{-1}}{2h} \\
\text{Forward:} \quad f'(x_0) &\approx \frac{-3f_0 + 4f_1 - f_2}{2h} \\
\text{Backward:} \quad f'(x_0) &\approx \frac{3f_0 - 4f_{-1} + f_{-2}}{2h}
\end{aligned}
$$

### Second Derivative

**3-point formulas:**
$$
\begin{aligned}
\text{Central:} \quad f''(x_0) &\approx \frac{f_{-1} - 2f_0 + f_1}{h^2} \\
\text{Forward:} \quad f''(x_0) &\approx \frac{f_0 - 2f_1 + f_2}{h^2} \\
\text{Backward:} \quad f''(x_0) &\approx \frac{f_0 - 2f_{-1} + f_{-2}}{h^2}
\end{aligned}
$$
